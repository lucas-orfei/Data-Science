```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
houses_uri = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro05/oficina/arquivos/dadosoficina05.txt'
houses = pd.read_csv(houses_uri)
houses.head()
```

Crie uma visualização que mostre a distribuição dos valores de aluguel, ‘rent amount (R$)’, dos imóveis. Pode utilizar o gráfico que você julgar mais adequado. Qual faixa de valores de aluguel é mais comum?

```python
plt.figure(figsize=(10, 5))
plt.title('Distribuição dos valores de aluguel dos imóveis')
sns.kdeplot(houses['rent amount (R$)'])
plt.show()
```

Portanto, os valores mais comuns estão no pico do gráfico: mais ou menos R$ 2500,00

Crie uma visualização que mostre a distribuição dos valores de seguro contra incêndio, 'fire insurance (R$)', dos imóveis. Existem valores incomuns?
