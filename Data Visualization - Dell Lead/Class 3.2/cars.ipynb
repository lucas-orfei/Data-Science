```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
mpg = sns.load_dataset("mpg")
mpg.head()
```

Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível diferenciar o país de origem (‘origin’). Sabendo disso, existe uma correlação entre a aceleração e o consumo? De que tipo? A correlação é a mesma para os diferentes países de origem? Você pode escolher a técnica que julgar mais adequada!

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre aceleração e o consumo de combustível (mpg)')
sns.scatterplot(x="acceleration", y="mpg", hue = "origin", data = mpg)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre aceleração e o consumo de combustível (mpg)')
sns.relplot(x="acceleration", y="mpg", col = "origin", data = mpg)
plt.show()
```

Portanto, vemos que existe uma relação positiva entre a aceleração e o consumo de combustível em todos os países.

Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível visualizar o ano do modelo (model_year). Existe alguma relação entre as três variáveis?

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre aceleração e o consumo de combustível (mpg)')
sns.scatterplot(x="acceleration", y="mpg", hue = "model_year", data = mpg)
plt.show()
```

Portanto, vemos que conforme os anos foram passando o consumo de combustível aumentou consideravalmente para uma mesma aceleração.
