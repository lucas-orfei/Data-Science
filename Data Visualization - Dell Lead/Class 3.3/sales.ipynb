```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
sales = pd.read_csv('https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro07/oficina/arquivos/dadosoficina07.txt')
sales.head()
```

Crie uma visualização que mostre a série temporal das vendas ao longo dos meses. Você pode, por exemplo, se perguntar se existe alguma tendência ou alguma periodicidade.

```python
plt.figure(figsize=(10,5))
plt.title('Vendas ao longo dos meses')
sns.lineplot(x="Month", y="Sales", data=sales)
plt.show()
```

Existe uma tendência das vendas aumentarem com os meses.
