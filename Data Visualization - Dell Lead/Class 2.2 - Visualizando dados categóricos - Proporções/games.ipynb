```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
games_uri = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro04/oficina/arquivos/dadosoficina04.txt'
games = pd.read_csv(games_uri)
games.head()
```

Crie uma visualização que mostre a proporção de vendas global entre gêneros de jogos da empresa Nintendo.

Vamos criar um novo DataFrame que mostre todos os jogos e vendas da empresa Nintendo.

```python
nintendo_games = games[games['Publisher'] == 'Nintendo']
nintendo_games
```

Agora, com esse novo DataFrame, podemos agrupar os gêneros dos jogos produzidos pela Nintendo.

```python
generos = nintendo_games.groupby(by='Genre').sum()
generos
```

Dessa forma, podemos plotar um gráfico de barras que mostre a porcentagem de vendas de jogos por cada gênero.

```python
generos = generos.sort_values(by='Global_Sales',ascending=False)
porcentagens = generos['Global_Sales']*100.0/generos['Global_Sales'].sum()
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(15,5))
plt.bar(generos.index,porcentagens)
plt.title('Proporção de vendas de jogos por gênero da Nintendo')
plt.show()
```

Crie uma visualização que mostre a proporção de quantidade de jogos por plataforma a partir do ano de 2015.

Vamos criar um novo DataFrame que filtre todos os jogos produzidos a partir de 2015.

```python
games_2015= games[games['Year'] >= 2015.0]
games_2015
```

Podemos, agora, agrupar todos os jogos produzidos a partir de 2015 por plataforma.

```python
plataformas = games_2015.groupby(by='Platform').count()
plataformas
```

```python
porcentagens = plataformas['Global_Sales']*100.0/plataformas['Global_Sales'].sum()
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(15,5))
plt.bar(plataformas.index,porcentagens)
plt.title('Proporção de jogos a partir de 2015 por plataformas')
plt.show()
```
