```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

Crie uma visualização que mostre a distribuição do valor da passagem (“fare”), de acordo com a cidade na qual o passageiro embarcou ("embark_town"). Você pode escolher a técnica que julgar mais adequada;

```python
plt.figure(figsize=(10,5))
plt.title('Distribuição do valor da passagem em função da cidade de embarque')
sns.violinplot(x="embark_town", y="fare", data=titanic)
plt.show()
```

É perceptível que mais pessoas da Queenstown pagaram valores menores em relação com outras cidades. Vemos ainda que Cherbourg representa a cidade onde mais pessoas pagaram por valores maiores na passagem. Enquanto Southampton fica no meio termo entre as duas outras cidades.

Crie uma visualização que mostre as distribuições do valor da passagem, de acordo com a cidade de embarque, sexo e se o passageiro sobreviveu ou não. Você pode escolher a técnica que julgar mais adequada. Que diferença você consegue notar nas distribuições dos passageiros que sobreviveram em relação aos que não sobreviveram?

```python
sns.catplot(x="embark_town", y="fare", hue = 'sex', col = 'survived', kind = 'violin', split = True, data=titanic)
plt.show()
```

```python
sns.catplot(x="embark_town", y="fare", hue = 'sex', col = 'survived', kind = 'box', data=titanic)
plt.show()
```

Vemos que mais mulheres sobreviveram em relação aos homens e além disso, as pessoas de Cherbourg sobreviveram mais do que as outras cidades.
